```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
#warnings.simplefilter(action = "ignore", category = FutureWarning)
```

```python
matches = pd.read_csv('dataset/dataset1_test.csv') 
```

```python
Records=pd.DataFrame(matches)
Records.columns=['Batsman','Runs','Minute','Ballfaced','fours','sixes','StrikeRate','Position','DismissalAt','Innings']
#Records.head(10)
```

```python
# split individual player records

batsman1 = Records[(Records.Batsman==1)].reset_index()
batsman2 = Records[(Records.Batsman==2)].reset_index()
batsman3 = Records[(Records.Batsman==3)].reset_index()
batsman4 = Records[(Records.Batsman==4)].reset_index()
batsman5 = Records[(Records.Batsman==5)].reset_index()
batsman6 = Records[(Records.Batsman==6)].reset_index()
```

```python
#batsman run by four sixes comparison

runBysixFour= Records[['Batsman','fours','sixes','Runs']].groupby('Batsman').sum() #run from boundary from each batsman
runBysixFour['totalRunFromBoundary']=runBysixFour['fours']*4+runBysixFour['sixes']*6
runBysixFour['runByFour']=runBysixFour['fours']*4
runBysixFour['runBysix']=runBysixFour['sixes']*6
runBysixFour['totalRunFromBoundary']
print(runBysixFour)

plt.rcParams["figure.figsize"] = [10,7]
plt.show(runBysixFour.plot.bar())
```

```python
#T-Distributio of Batsman1 with 95% confidence level

#Sample Mean,x
#Sample Standard Deviation,s
#Sample size, n
#Degree of Freedom, df=n-1
#alpha level,alpha
#confidence level, cl

import math

print('batsman 1 : mean ',batsman1_runs.mean(),' Standard Deviation ',batsman1_runs.std())
batsman1_x = batsman1_runs.mean();
batsman1_s = batsman1_runs.std();
batsman1_n = batsman1_runs.count();#samle size
batsman1_df = batsman1_n - 1;
batsman1_cl = 0.95; #98%
batsman1_alpha = (1-batsman1_cl)/2; #divide by 2 because its 1-Tail Test.

print('Batsman 1 Alpha Level',batsman1_alpha,'\n')
print('Batsman 1 Degree of Freedom',batsman1_df,'\n')
```

```python
print('Since Df is 56 we take closest value which is 60')
batsman1_t_score = 2.000
batsman1_CI_upperLevel = batsman1_x + batsman1_t_score*(batsman1_s/math.sqrt(batsman1_n))
batsman1_CI_lowerLevel = batsman1_x - batsman1_t_score*(batsman1_s/math.sqrt(batsman1_n))

print('Batsman 1 Confidence Intervel UpperLevel',batsman1_CI_upperLevel,'\n')
print('Batsman 1 Confidence Intervel LowerLevel',batsman1_CI_lowerLevel,'\n')
```
